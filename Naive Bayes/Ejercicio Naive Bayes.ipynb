# Naive Bayes Classification

```python
# Importamos los datos
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

Los datos de imagen son una array tridimensional: 1797 muestras, cada uno con una cuadricula de pixeles de 8x8

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(10, 10, figsize=(8, 8), 
                         subplot_kw={'xticks':[], 'yticks':[]}, 
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
            transform=ax.transAxes, color='green')
```

```python
# Variables
X = digits.data
X.shape
```

```python
# Variable dependiente
y = digits.target
y.shape
```

## Gaussian Naive Bayes Model

```python
# Dividimos los datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, train_size=0.75 ,random_state=0)
```

```python
# Ajustamos el modelo
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
```

```python
# Realizamos predicciones usando los datos de prueba
y_model = model.predict(Xtest)
```

```python
# Revisamos el accuracy 
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
# Matriz de confusin
import seaborn as sns
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, y_model)
sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```

## Multinomial Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model2 = MultinomialNB()
model2.fit(Xtrain, ytrain)
```

```python
# Realizamos predicciones usando los datos de prueba
y_model2 = model2.predict(Xtest)
```

```python
# Revisamos el accuracy 
accuracy_score(ytest, y_model2)
```

```python
# Matriz de confusin

mat2 = confusion_matrix(ytest, y_model2)
sns.heatmap(mat2, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```
